**Scenario Design Analysis for Yelp**

**Target User** 

    Local business, general population


**Key goals** 

    To serve the people/customers by providing them to discover,connect and transact with local businesses.



**How to accomplish the goal ?**
    
    
    In order to accomplish the goal it is important to attract the high-intent customers by showing them the best choice.Businesses must claim their business because by doing so they will  have access to additional page features and tools to add the most up-to-date information to the business page. Adding photos with captions will help to layout the information to the customers.In addition to that photo captions will allow the business to get more visibility in search results. Responding to the reviews & community questions,updating and including all the information regarding bussiness is really important. For example, updating business hours and operation on holidays and using promote features will really help and be effective to reach and stay connected with the customers.


**Reverse engineering**
    
        I noticed that the results of the same search varries from which platform you search. Search results on web and mobile is different. I think providing the exact results is really important and this must be emphasized. I also believe that user experience would have been much better if  tabs were provided for the food cuisine when searching up delivery as well. Yelp now only gives those tabs when clicked on "Restaurants". It is all about improving the user convenience.Therefore, working on these area will definitely help to improve the user experience.
